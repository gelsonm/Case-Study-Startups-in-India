```python
#Que_1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
file=pd.read_csv("Datasets\startup_funding.csv",encoding = 'utf-8')
file["CityLocation"].dropna(inplace = True)
def splitcity(s):
    return s.split("/")[0].strip()
file['CityLocation'] = file['CityLocation'].astype(str).apply(splitcity)

file['CityLocation'].replace("Delhi","New Delhi", inplace = True)
file['CityLocation'].replace("bangalore","Bangalore", inplace = True)

file1 = file[(file["CityLocation"] == "Bangalore") | (file["CityLocation"] == "New Delhi") | (file["CityLocation"] == "Mumbai")| (file["CityLocation"] == "Gurgaon" )| (file["CityLocation"] == "Noida")]

city_fundings = file1["CityLocation"].value_counts()[0:5]
city_names = city_fundings.index
fundings = city_fundings.values
print('Location List with No of Fundings')
for i in range(len(city_names)):
    print(city_names[i], fundings[i])
print('Location having maximum number of fundings:'+city_names[0])
print
plt.bar(city_names,fundings, width = 0.4)
plt.title("City Versus No. of Fundings")
plt.xlabel('City')
plt.ylabel('No of Fundings Received')
plt.show()
```

```python
# Que-2

import numpy as np
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import csv
df = pd.read_csv("Datasets\startup_funding.csv",encoding = 'utf-8')
df = df.dropna(subset=['InvestorsName'])
d={}
fundings = []
investors_name = []
def separateInvestors(n):    
    n=n.split(',')
    l=[]
    
    for i in n:
        l.append(i.strip())
    for i in l:
        if i in d:
            d[i]=d[i]+1
        else:
            d[i]=1
df['InvestorsName']=df['InvestorsName'].apply(separateInvestors)
r = Counter(d)
investors = r.most_common(5)  
print('Location List with No of Fundings')
for i in range(len(investors)): 
    fundings.append(investors[i][1])
    investors_name.append(investors[i][0])
    print(investors[i][0],investors[i][1]) 
plt.bar(investors_name,fundings ,width = 0.4)
plt.title("Investors Versus No. of Fundings")
plt.xlabel('Investors')
plt.ylabel('No of Fundings')
plt.xticks(rotation ='vertical')
plt.show()
```

```python
#que-3

import numpy as np
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import csv
df_start = pd.read_csv("Datasets\startup_funding.csv",encoding = 'utf-8')
df_start.dropna(subset=['InvestorsName','StartupName'],inplace = True)
df_start['StartupName'].replace("ola","Ola", inplace = True)
df_start['StartupName'].replace("Ola Cabs","Ola", inplace = True)
df_start['StartupName'].replace("olacab","Ola", inplace = True)
df_start['StartupName'].replace("Olacabs","Ola", inplace = True)
df_start['StartupName'].replace("OYO Rooms","Oyo", inplace = True)
df_start['StartupName'].replace("Oyo Rooms","Oyo", inplace = True)
df_start['StartupName'].replace("OyoRooms","Oyo", inplace = True)
df_start['StartupName'].replace("Oyorooms","Oyo", inplace = True)
df_start['StartupName'].replace("OYO Rooms","Oyo", inplace = True)
df_start['StartupName'].replace("Paytm Marketplace","Paytm", inplace = True)
df_start['StartupName'].replace("Flipkart.com","Flipkart", inplace = True)
d = {}

investments = []
investors_name = []
count = 1
startup = list(df_start['StartupName'])
investors = list(df_start['InvestorsName'])

print('Location List with No of Fundings')
for i in range(len(investors)):
    investor = investors[i].split(',')
    for invest in investor:
        invest = invest.strip()
        if invest != "" and (invest != "Undisclosed Investors"):
            if invest in d:
                s = d[invest]
                s.add(startup[i])
                d[invest] = s
                
            else:
                d[invest] = {startup[i]}
for ele in d:
    d[ele] = len(d[ele])

d_sorted = {k: v for k, v in sorted(d.items(),key = lambda item: item[1], reverse = True)}

for ele in d_sorted:
    if count > 5:
        break
    investors_name.append(ele)
    investments.append(d_sorted[ele])
    print(ele,d_sorted[ele])
    count += 1


plt.bar(investors_name,investments ,width = 0.4)
plt.title("Investors Versus No. of Fundings")
plt.xlabel('Investors')
plt.ylabel('No of Fundings')
plt.xticks(rotation ='vertical')
plt.show()          
        

```
